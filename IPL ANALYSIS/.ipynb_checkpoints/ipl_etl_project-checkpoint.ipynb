IMPORT LIBRARIES

```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
```

EXTRACT – LOAD CSV FILES

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
players = pd.read_csv("players.csv")

matches.head(), deliveries.head(), players.head()
```

```python
print(matches.shape)
print(deliveries.shape)
print(players.shape)

matches.info()
```

TRANSFORM – DATA CLEANING

```python
matches.isnull().sum(), deliveries.isnull().sum(), players.isnull().sum()
```

```python
matches = matches.drop_duplicates()
deliveries = deliveries.drop_duplicates()
players = players.drop_duplicates()
```

```python
matches["winner"] = matches["winner"].fillna("No Result")
matches["venue"] = matches["venue"].fillna("Unknown")
deliveries[["batsman_runs", "bowler_runs", "is_wicket"]] = (deliveries[["batsman_runs", "bowler_runs", "is_wicket"]].fillna(0))
players["role"] = players["role"].fillna("Unknown")
```

TRANSFORM – BUSINESS AGGREGATIONS

```python
deliveries["total_runs"] = (deliveries["batsman_runs"] + deliveries["bowler_runs"])
```

```python
runs_per_match = (deliveries.groupby("match_id")["total_runs"].sum().reset_index(name="match_runs"))

runs_per_match.head()
team_wins = (matches[matches["winner"] != "No Result"].groupby("winner").size().reset_index(name="wins"))

team_wins.head()
wickets_per_match = (deliveries .groupby("match_id")["is_wicket"].sum().reset_index(name="wickets"))

wickets_per_match.head()
```

CONNECT TO MYSQL

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:15625@localhost:3306/ipl_db")
engine
```

```python
matches.to_sql(
    name="matches",
    con=engine,
    if_exists="replace",
    index=False
)

deliveries.to_sql(
    name="deliveries",
    con=engine,
    if_exists="replace",
    index=False
)

players.to_sql(
    name="players",
    con=engine,
    if_exists="replace",
    index=False
)

print(" Raw tables loaded")
```

```python
runs_per_match.to_sql(
    name="runs_per_match",
    con=engine,
    if_exists="replace",
    index=False
)

team_wins.to_sql(
    name="team_wins",
    con=engine,
    if_exists="replace",
    index=False
)

wickets_per_match.to_sql(
    name="wickets_per_match",
    con=engine,
    if_exists="replace",
    index=False
)

print(" Analytics tables loaded")
```

Dashboards

Dashboard-1

```python
df_team_highest_wins = pd.read_sql("""
SELECT winner AS team, COUNT(*) AS total_wins
FROM matches
WHERE winner != 'No Result'
GROUP BY winner
ORDER BY total_wins DESC
""", engine)

df_team_highest_wins.to_csv("dashboard1_team_highest_wins.csv", index=False)
```

```python
df_season_team_dominance = pd.read_sql("""
SELECT season, winner AS team, COUNT(*) AS wins
FROM matches
WHERE winner != 'No Result'
GROUP BY season, winner
ORDER BY season, wins DESC
""", engine)

df_season_team_dominance.to_csv("dashboard1_season_team_dominance.csv", index=False)
```

```python
df_wins_by_venue = pd.read_sql("""
SELECT venue, winner AS team, COUNT(*) AS wins
FROM matches
WHERE winner != 'No Result'
GROUP BY venue, winner
ORDER BY venue, wins DESC
""", engine)

df_wins_by_venue.to_csv("dashboard1_wins_by_venue.csv", index=False)
```

Dashboard-2

```python
df_avg_runs_per_match = pd.read_sql("""
SELECT AVG(match_runs) AS avg_runs_per_match
FROM runs_per_match
""", engine)

df_avg_runs_per_match.to_csv("dashboard2_avg_runs_per_match.csv", index=False)
```

```python
df_no_result_matches = pd.read_sql("""
SELECT COUNT(*) AS no_result_matches
FROM matches
WHERE winner = 'No Result'
""", engine)

df_no_result_matches.to_csv("dashboard2_no_result_matches.csv", index=False)
```

```python
df_home_vs_away = pd.read_sql("""
SELECT 
    winner AS team,
    SUM(CASE WHEN winner = team1 THEN 1 ELSE 0 END) AS home_wins,
    SUM(CASE WHEN winner = team2 THEN 1 ELSE 0 END) AS away_wins
FROM matches
WHERE winner != 'No Result'
GROUP BY winner
ORDER BY home_wins DESC, away_wins DESC
""", engine)

df_home_vs_away.to_csv("dashboard2_home_vs_away_performance.csv", index=False)
```

Dashboard-3

```python
df_runs_distribution = pd.read_sql("""
SELECT match_id, match_runs
FROM runs_per_match
ORDER BY match_runs DESC
""", engine)

df_runs_distribution.to_csv("dashboard3_runs_distribution_per_match.csv", index=False)
```

```python
df_wickets_per_match = pd.read_sql("""
SELECT match_id, wickets
FROM wickets_per_match
ORDER BY wickets DESC
""", engine)

df_wickets_per_match.to_csv("dashboard3_wickets_per_match.csv", index=False)
```

```python
df_high_scoring_matches = pd.read_sql("""
SELECT match_id, match_runs
FROM runs_per_match
WHERE match_runs > (
    SELECT AVG(match_runs) FROM runs_per_match
)
ORDER BY match_runs DESC
""", engine)

df_high_scoring_matches.to_csv("dashboard3_high_scoring_matches.csv", index=False)
```

Dashboard-4

```python
df_top5_win_percentage = pd.read_sql("""
SELECT 
    winner AS team,
    COUNT(*) * 100.0 / (SELECT COUNT(*) FROM matches WHERE winner != 'No Result')
        AS win_percentage
FROM matches
WHERE winner != 'No Result'
GROUP BY winner
ORDER BY win_percentage DESC
LIMIT 5
""", engine)

df_top5_win_percentage.to_csv("dashboard4_top5_teams_by_win_percentage.csv", index=False)
```

```python
df_venue_impact = pd.read_sql("""
SELECT venue, winner AS team, COUNT(*) AS wins
FROM matches
WHERE winner != 'No Result'
GROUP BY venue, winner
ORDER BY venue, wins DESC
""", engine)

df_venue_impact.to_csv("dashboard4_venue_impact_on_results.csv", index=False)
```

```python
venue_impact_total = df_venue_impact.groupby("venue")["wins"].sum()

plt.figure(figsize=(12,5))
plt.bar(venue_impact_total.index, venue_impact_total.values)
plt.xticks(rotation=45)
plt.title("Impact of Venue on Match Results")
plt.xlabel("Venue")
plt.ylabel("Wins")
plt.show()
```
